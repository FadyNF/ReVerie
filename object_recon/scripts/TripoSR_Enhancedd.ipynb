```python

import os
from IPython.display import Video, display, HTML

out_dir = 'output'
print('Contents of output dir:', os.listdir(out_dir) if os.path.exists(out_dir) else 'No output directory yet.')

mp4 = os.path.join(out_dir, 'render.mp4')
if os.path.exists(mp4):
    display(Video(mp4, embed=True, width=640))
else:
    print('No render.mp4 found.')

meshes = [f for f in os.listdir(out_dir)] if os.path.exists(out_dir) else []
print('Output files:', meshes)

for fname in meshes:
    display(HTML(f"<a href='/content/TripoSR/{out_dir}/{fname}' target='_blank' download>{fname}</a>"))

```
